```python
# 1. Extract QR Code Value from left side box [Done]
def read_qr_code(filename):
    """Read an image and read the QR code.
    Args:
        filename (string): Path to file
    Returns:
        qr (string): Value from QR code
    """
    try:
        img = cv2.imread(filename)
        detect = cv2.QRCodeDetector()
        value, points, straight_qrcode = detect.detectAndDecode(img)
        return value
    except:
        return "Error in reading QR CODE!"
```

```python
# 3. Extract the total value from right side table + OMR [Done]

# https://stackoverflow.com/questions/57964634/python-opencv-skew-correction-for-ocr
# import cv2
# import numpy as np
from scipy.ndimage import interpolation as inter

def correct_skew(image, delta=1, limit=5):
    def determine_score(arr, angle):
        data = inter.rotate(arr, angle, reshape=False, order=0)
        histogram = np.sum(data, axis=1, dtype=float)
        score = np.sum((histogram[1:] - histogram[:-1]) ** 2, dtype=float)
        return histogram, score

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1] 

    scores = []
    angles = np.arange(-limit, limit + delta, delta)
    for angle in angles:
        histogram, score = determine_score(thresh, angle)
        scores.append(score)

    best_angle = angles[scores.index(max(scores))]

    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, best_angle, 1.0)
    corrected = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_CUBIC, \
            borderMode=cv2.BORDER_REPLICATE)

    return best_angle, corrected

def deskew_img(filename):
    image = cv2.imread(filename)
    angle, corrected = correct_skew(image)
    print('Skew angle:', angle)
    # cv2.imwrite('output/corrected.jpg', corrected)
    return corrected
    
```

```python
# def getCornerPoints(omr_cont):
#     peri = cv2.arcLength(omr_cont,True)
#     approx = cv2.approxPolyDP(omr_cont, 0.02*peri, True)
#     return approx

def reorder_indexes(myPts):
    myPts = myPts.reshape((4,2))
    myPtsNew = np.zeros((4,1,2), np.int32)
    add = myPts.sum(1)
    myPtsNew[0] = myPts[np.argmin(add)] # [0,0]
    myPtsNew[3] = myPts[np.argmax(add)] # [w,h]
    
    diff = np.diff(myPts, axis =1)
    myPtsNew[1] = myPts[np.argmin(diff)] # [w,0]
    myPtsNew[2] = myPts[np.argmax(diff)] # [0, h]
    
    return myPtsNew
    
def splitCellsOMR(omrImg):
    rows = np.vsplit(omrImg, 10)
    cells = []
    for c in rows:
        cols = np.hsplit(c, 2)
        for cell in cols:
            cells.append(cell)
            # cv2.imwrite('output/11.jpg', cell)
    return cells

def getMaxArgs(cells):
    arr = []
    for cell in cells:
        arr.append(cv2.countNonZero(cell))
    arg1 = 0
    arg2 = 0
    for i in range(0,len(arr),2):
        if arr[i] > arg1:
            arg1 = arr[i]
            ind1 = i//2

    for i in range(1,len(arr),2):
        if arr[i] > arg2:
            arg2 = arr[i]
            ind2 = i//2

    return [(ind1,ind2) , (arg1,arg2)]
    
```
